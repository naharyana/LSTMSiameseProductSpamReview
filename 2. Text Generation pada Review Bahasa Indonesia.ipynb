```
from transformers import GPT2Tokenizer, TFGPT2Model

model_name='cahya/gpt2-small-indonesian-522M'
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = TFGPT2Model.from_pretrained(model_name)
```

```
text = "Silakan diganti dengan text apa saja."
encoded_input = tokenizer(text, return_tensors='tf')
output = model(encoded_input)
```

```
from transformers import GPT2TokenizerFast, GPT2LMHeadModel
pretrained_name = "w11wo/indo-gpt2-small"
tokenizer = GPT2TokenizerFast.from_pretrained(pretrained_name)
tokenizer.model_max_length = 1024
model = GPT2LMHeadModel.from_pretrained(pretrained_name)
```

```
# sample prompt
prompt = "Nama saya Budi, dari Indonesia"
input_ids = tokenizer.encode(prompt, return_tensors='pt')
model.eval()

# generate output using top-k sampling
sample_outputs = model.generate(input_ids, 
                                pad_token_id=50256,
                                do_sample=True, 
                                max_length=40, 
                                min_length=40,
                                top_k=40,
                                num_return_sequences=1)

for i, sample_output in enumerate(sample_outputs):
    print(tokenizer.decode(sample_output.tolist()))
```

```
import pandas as pd
rev = pd.read_csv("/content/drive/MyDrive/LSTMSiamese/data/revLazCleanMin30words.csv")
```

```
lines=[]
for i in rev['reviewClean']:
  lines.append(i)
```

```
seed=[]

for i in lines:
    # select a seed text
    seed_text = i #randint(0,len(i)/2)
    text = seed_text.split()
    n = int(len(text)/2)+1
    text_split=[' '.join(text[i:i+n]) for i in range(0,len(text),n)]
    seed.append(text_split)
    #print(text_split[0] + '\n')
```

```
prompts = [i[0] for i in seed]
```

```
for i in prompts:
  print(i)
```

```
prompts[1:4]
```

```
# sample prompt
gen=[]
for prompt in prompts[1:4]:
  input_ids = tokenizer.encode(prompt, return_tensors='pt')
  model.eval()
  # generate output using top-k sampling
  sample_outputs = model.generate(input_ids, 
                                pad_token_id=50256,
                                do_sample=True, 
                                max_length=40, 
                                min_length=40,
                                top_k=40,
                                num_return_sequences=1)
  for i, sample_output in enumerate(sample_outputs):
    print(tokenizer.decode(sample_output.tolist()))
    gen.append(tokenizer.decode(sample_output.tolist()))
```
